```
#preprocessing of data
# 1. resize
# 2. flatten

import os
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize

target=[]
images=[]
flat_data=[]

DATADIR='/content/drive/MyDrive/images'
CATEGORIES=['ice cream cone','rugby ball leather','sunflower']

for category in CATEGORIES:
  class_num=CATEGORIES.index(category) # label encoding for categories
  # print(class_num)
  # output=[0 1 2]
  path=os.path.join(DATADIR,category)  # create path to use all the images
  # print(path)
  # /content/images/Dog
  # /content/images/Cat
  # /content/images/Goat
  for img in os.listdir(path):
    img_array=imread(os.path.join(path,img))
    # print( img_array.shape)
    # plt.imshow(img_array)
    # break
    img_resized=resize(img_array,(150,150,3)) # resizing the images to 150*150, resizing is also normalizing(scaling down) the pixel values between 0-1
    flat_data.append(img_resized.flatten()) # flattening the image from 2d to 1d
    images.append(img_resized)
    target.append(class_num)

flat_data=np.array(flat_data)
target=np.array(target)
images=np.array(images)
# /content/drive/MyDrive/images/ice cream cone?
```

```
unique,count=np.unique(target,return_counts=True)
plt.bar(CATEGORIES,count)
```

```
# splitting the data in training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(flat_data,target,test_size=0.3,random_state=109)
```

```
x_train
```

```
from sklearn.model_selection import GridSearchCV
from sklearn import svm
param_grid=[
            {'C':[1,10,100,1000],'kernel':['linear']},
            {'C':[1,10,100,1000],'gamma':[0.001,0.0001,0.00001],'kernel':['rbf']},
]
svc=svm.SVC(probability=True)
clf=GridSearchCV(svc,param_grid)
clf.fit(x_train,y_train)
```

```
y_pred=clf.predict(x_test)
y_pred
```

```
y_test
```

```
from sklearn.metrics import accuracy_score,confusion_matrix
```

```
accuracy_score(y_pred,y_test)*100
```

```
confusion_matrix(y_pred,y_test)
```

```
## saving the model using the picle lin=brary
import pickle
pickle.dump(clf,open('img_model.p','wb')) # writing the model to img_model.p file
```

```
model=pickle.load(open('img_model.p','rb')) # loading the model from img_model.p
```
